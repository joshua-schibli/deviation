IO, HT, St

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

current_directory = os.getcwd()

# Load the CSV file
filepath1 = f'{current_directory}\\fetchedData\\IO,HT,St_1.csv'
filepath2 = f'{current_directory}\\fetchedData\\IO,HT,St_2.csv'
filepath3 = f'{current_directory}\\fetchedData\\IO,HT,St_3.csv'
df1 = pd.read_csv(filepath1, sep=';')
df2 = pd.read_csv(filepath2, sep=';')
df3 = pd.read_csv(filepath3, sep=';')

std_deviation1 = df1[['XTran1', 'YTran1', 'ZTran1']].std()
total_deviation1 = (std_deviation1**2).sum() ** 0.5

std_deviation2 = df2[['XTran1', 'YTran1', 'ZTran1']].std()
total_deviation2 = (std_deviation2**2).sum() ** 0.5

std_deviation3 = df3[['XTran1', 'YTran1', 'ZTran1']].std()
total_deviation3 = (std_deviation3**2).sum() ** 0.5

print(f"deviation 1 between axes in (mm): \n{std_deviation1 * 1000}")
print(f"total deviation 1: {total_deviation1 * 1000}mm\n")

print(f"deviation 2 between axes in (mm): \n{std_deviation2 * 1000}")
print(f"total deviation 2: {total_deviation2 * 1000}mm\n")

print(f"deviation 3 between axes in (mm): \n{std_deviation3 * 1000}")
print(f"total deviation 3: {total_deviation3 * 1000}mm\n")
```

```python
filepath7 = f'{current_directory}\\fetchedData\\IO,HT,X_1.csv'
filepath8 = f'{current_directory}\\fetchedData\\IO,HT,X_2.csv'
filepath9 = f'{current_directory}\\fetchedData\\IO,HT,X_3.csv'
df7 = pd.read_csv(filepath7, sep=';')
df8 = pd.read_csv(filepath8, sep=';')
df9 = pd.read_csv(filepath9, sep=';')

filtered_min_data7 = df7[(df7['TotalTime'] >= 16.5) & (df7['TotalTime'] <= 20.3)]
filtered_max_data7 = df7[(df7['TotalTime'] >= 21.2) & (df7['TotalTime'] <= 25.4)]
filtered_line_data7 = df7[(df7['TotalTime'] >= 19.5)]
median_min_xtran7 = filtered_min_data7['XTran1'].median()
median_max_xtran7 = filtered_max_data7['XTran1'].median()
max_xtran7 = filtered_min_data7['XTran1'].max()
min_xtran7 = filtered_max_data7['XTran1'].min()

deviation7_1 = abs(median_max_xtran7 - median_min_xtran7)
deviation7_2 = abs(max_xtran7 - min_xtran7)

filtered_min_data8 = df8[(df8['TotalTime'] >= 11.17) & (df8['TotalTime'] <= 12.85)]
filtered_max_data8 = df8[(df8['TotalTime'] >= 13.8) & (df8['TotalTime'] <= 14.5)]
filtered_line_data8 = df8[(df8['TotalTime'] >= 14.7)]
median_min_xtran8 = filtered_min_data8['XTran1'].median()
median_max_xtran8 = filtered_max_data8['XTran1'].median()
min_xtran8 = filtered_line_data8['XTran1'].min()
max_xtran8 = filtered_line_data8['XTran1'].max()

deviation8_1 = abs(median_max_xtran8 - median_min_xtran8)
deviation8_2 = abs(max_xtran8 - min_xtran8)

filtered_min_data9 = df9[(df9['TotalTime'] >= 8.7) & (df9['TotalTime'] <= 11.2)]
filtered_max_data9 = df9[(df9['TotalTime'] >= 18.2) & (df9['TotalTime'] <= 22.6)]
filtered_line_data9 = df9[(df9['TotalTime'] >= 23)]
median_min_xtran9 = filtered_min_data9['XTran1'].median()
median_max_xtran9 = filtered_max_data9['XTran1'].median()
min_xtran9 = filtered_line_data9['XTran1'].min()
max_xtran9 = filtered_line_data9['XTran1'].max()

deviation9_1 = abs(median_max_xtran9 - median_min_xtran9)
deviation9_2 = abs(max_xtran9 - min_xtran9)

print(f"median 1: {deviation7_1}")
print(f"max 1: {deviation7_2}")
print(f"deviation 1: {abs(deviation7_2 - deviation7_1) * 1000}mm\n")

print(f"median 2: {deviation8_1}")
print(f"max 2: {deviation8_2}")
print(f"deviation 2: {abs(deviation8_2 - deviation8_1) * 1000}mm\n")

print(f"median 3: {deviation9_1}")
print(f"max 3: {deviation9_2}")
print(f"deviation 3: {abs(deviation9_2 - deviation9_1) * 1000}mm\n")

fig,ax = plt.subplots(nrows=3,ncols=1,figsize=(12,6))

# Hinzufügen von Titel und Achsenbeschriftungen
ax[0].plot(filtered_line_data7['XTran1'], linestyle='-')
ax[0].set_title('XTran1 über TotalTime')
ax[0].set_xlabel('TotalTime')
ax[0].set_ylabel('XTran1')

ax[1].plot(filtered_line_data8['XTran1'], linestyle='-')
ax[1].set_title('XTran1 über TotalTime')
ax[1].set_xlabel('TotalTime')
ax[1].set_ylabel('XTran1')

ax[2].plot(filtered_line_data9['XTran1'], linestyle='-')
ax[2].set_title('XTran1 über TotalTime')
ax[2].set_xlabel('TotalTime')
ax[2].set_ylabel('XTran1')

plt.tight_layout()
```
