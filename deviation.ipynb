```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

current_directory = os.getcwd()

# Load the CSV file
filepath1 = f'{current_directory}\\fetchedData\\IO,HT,St_1.csv'
filepath2 = f'{current_directory}\\fetchedData\\IO,HT,St_2.csv'
filepath3 = f'{current_directory}\\fetchedData\\IO,HT,St_3.csv'
df1 = pd.read_csv(filepath1, sep=';')
df2 = pd.read_csv(filepath2, sep=';')
df3 = pd.read_csv(filepath3, sep=';')

std_deviation1 = df1[['XTran1', 'YTran1', 'ZTran1']].std()
total_deviation1 = (std_deviation1**2).sum() ** 0.5

std_deviation2 = df2[['XTran1', 'YTran1', 'ZTran1']].std()
total_deviation2 = (std_deviation2**2).sum() ** 0.5

std_deviation3 = df3[['XTran1', 'YTran1', 'ZTran1']].std()
total_deviation3 = (std_deviation3**2).sum() ** 0.5

print(f"deviation 1 between axes in (mm): \n{std_deviation1 * 1000}")
print(f"total deviation 1: {total_deviation1 * 1000}mm\n")

print(f"deviation 2 between axes in (mm): \n{std_deviation2 * 1000}")
print(f"total deviation 2: {total_deviation2 * 1000}mm\n")

print(f"deviation 3 between axes in (mm): \n{std_deviation3 * 1000}")
print(f"total deviation 3: {total_deviation3 * 1000}mm\n")
```

```python
# Define a function to multiply the tick value by 1000
def multiply_by_1000(x, pos):
    return '{:.0f}'.format(x * 1000)
```
